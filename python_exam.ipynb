```python
# gorunduyu kimi bu sutunlar arasinda elaqe hec yox deylecek qederdir. yeni mehsulun qaranti muddeti onun qiymetine tesir etmir. bir mehsulum qaranti
# muddeti 1 olmasi ve ya 3 olmasi onun qiymetini deyismir ve ya mehsullarin qiymetinin cox olmasi onun stokda daha cox ve ya az olmasina da tesir 
#etmir. 
```

```python
#sual2
en=[]
uzunluq=[]
hundurluk=[]
for i in range(df3.shape[1]):
    en.append(df3.loc[i,"Product Dimensions"][:df3.loc[i,"Product Dimensions"].index("x")])
    uzunluq.append(df3.loc[i,"Product Dimensions"][df3.loc[i,"Product Dimensions"].index("x")+1:
                   df3.loc[i,"Product Dimensions"].index("x",df3.loc[i,"Product Dimensions"].index("x")+1)])
    hundurluk.append(df3.loc[i,"Product Dimensions"][df3.loc[i,"Product Dimensions"].index("x",df3.loc[i,"Product Dimensions"].index("x")+1)+1:df3.loc[i,"Product Dimensions"].index(" ")])
```

```python
df4 = pd.DataFrame({
    'en': en,
    'uzunluq': uzunluq,
    'hundurluk' : hundurluk})
df4["hecm"]=df4["en"].astype(int)*df4["uzunluq"].astype(int)*df4["hundurluk"].astype(int)
df3 = pd.concat([df3,df4], axis=1)
df3
```

```python
sns.scatterplot(df3,x="hecm",y="Price")
```

```python
df3[["Price","hecm"]].corr()
```
